```python
import pandas as pd

url_file1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url_file2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv"
url_file3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv"

df_file1 = pd.read_csv(url_file1)
df_file2 = pd.read_csv(url_file2)
df_file3 = pd.read_csv(url_file3)

print("DataFrame 1:")
print(df_file1.head())

print("\nDataFrame 2:")
print(df_file2.head())

print("\nDataFrame 3:")
print(df_file3.head())

customer_data = [['df_file1', 'df_file2', 'df_file3']]

print(customer_data)
```

```python
customer_data = pd.concat([df_file1, df_file2, df_file3], ignore_index=True)

print("\nCombined DataFrame:")
print(customer_data.head())
print(customer_data.shape)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data_path = "Data_Marketing_Customer_Analysis_Round3.csv"
df = pd.read_csv(data_path)

print("DataFrame_:")
print(df.head())

numerical = df.select_dtypes(np.number, object)
categorical = df.select_dtypes(object, np.number)



numerical.hist(figsize=(10,10), xrot=20)

numerical.income.hist(figsize=(10,10), xrot=20)

fig = plt.figure(figsize=(10,10))
sns.heatmap(numerical.corr(), annot=True)

response_df = df.groupby('response')['income'].count().reset_index()

plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
plt.title('number of responses', size='25', fontweight='bold')

sns.barplot(x = df['response'])

sns.countplot(x=df['sales_channel'], hue = df['response'])

df['claim_amount_group'] = df['total_claim_amount'].apply(lambda x: 'over 400 claims' if x > 400 else 'Equal or Less than 400')
sns.countplot(x=df['claim_amount_group'], hue = df['response'])

df['claim_amount_group'] = pd.cut(df['total_claim_amount'], 6)
g=sns.countplot(x=df['claim_amount_group'], hue = df['response'])
g.set_xticklabels(g.get_xticklabels(), rotation=20)

df['income_amount_group'] = pd.cut(df['income'], 10)
h=sns.countplot(x=df['income_amount_group'], hue = df['response'])
h.set_xticklabels(h.get_xticklabels(), rotation=20)
```
